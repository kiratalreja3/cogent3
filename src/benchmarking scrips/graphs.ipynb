```python
import seaborn as sns
sns.set_style("whitegrid")
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {'New Approach':[152.5,631.7,711.8]}
df = pd.DataFrame(data)
fig, ax = plt.subplots() 
m = sns.lineplot(df,ax=ax,markers=True,linewidth = 3)
ax.set_xticks(range(0,3));
plt.legend(title='Memory Usage');  
m.set_xticklabels(['Func Start','Load Seq','Annotate & Query'])
ax.set(xlabel='Operation', ylabel='Memory in MiB')
ax.set_title('Memory Usage for GFF Annotations with the New Approach');
plt.savefig("Memory-NewApproach.svg", format="svg")
```

```python
old_data = {'Old Approach':[132.1,717.2,1422.2],'New Approach':[152.5,631.7,711.8]}

old_df = pd.DataFrame(old_data)

fig, ax = plt.subplots()
m = sns.lineplot(old_df,ax=ax,markers=True,linewidth = 3)

ax.set_xticks(range(0,3));
plt.legend(title='Memory Usage');  
m.set_xticklabels(['Func Start','Load Seq','Annotate & Query'])
ax.set(xlabel='Operation', ylabel='Memory in MiB')
ax.set_title('Memory Usage for GFF Annotations');
ax.lines[1].set_linestyle("-")
plt.savefig("Memory-Comparison.svg", format="svg")
```

```python
data = {'Old Approach':[6.37,6.28,6.48,7.37,11.341,18.09,25,29.57]}
df = pd.DataFrame(data)
fig, ax = plt.subplots()
m = sns.lineplot(df,ax=ax,markers=True,linewidth = 3)
#ax.set_xticks(range(0,8));
plt.legend(title='Time Complexity of Old Approach');  
m.set_xticklabels(['500','5k','10k','20k','50k','75k','100k','125k','150k'])
ax.set(xlabel='Number of annotations', ylabel='Time taken')
ax.set_title('Time Complexity of GFF Annotations with Old Approach');
plt.savefig("TimeComplexity-OldApproach.svg", format="svg");
```

```python
data = {'Old Approach':[6.37,6.28,6.48,7.37,11.341,18.09,25,29.57,32],'New Approach':[5.28,5.38,5.48,6.12,6.38,6.39,6.41,6.51,8.28]}

old_df = pd.DataFrame(data)

fig, ax = plt.subplots()
m = sns.lineplot(old_df,ax=ax,markers=True,linewidth = 3)

plt.legend(title='Time Complexity');  
m.set_xticklabels(['0','500','5k','10k','20k','50k','75k','100k','125k','150k'])
ax.set(xlabel='Number of annotations', ylabel='Time')
ax.set_title('Time Complexity for GFF Annotations');
ax.lines[1].set_linestyle("-")
plt.savefig("TimeComplexity-Comparison.svg", format="svg")
```

```python
data = {'Time taken':[8.8,201.1],'Operations':['Load Seq','Annotate & Query']}
old_df = pd.DataFrame(data)
fig, ax = plt.subplots()
sns.barplot(data=old_df,x='Operations',y='Time taken',ax=ax);
ax.set_title('Time taken for GFF Annotations');
plt.legend(title='Old Approach'); 
plt.savefig("TimeProfile-OldApproach.svg", format="svg")
```

```python
data = {'Time taken':[8.8,201.1,9.23,11.03],'Operations':['Load Seq','Annotate & Query','Load Seq','Annotate & Query'],'Approach':['Old Approach','Old Approach','New Approach','New Approach']}
old_df = pd.DataFrame(data)
fig, ax = plt.subplots()
sns.barplot(data=old_df,x='Operations',y='Time taken',hue='Approach',ax=ax);
plt.legend(title='Approaches'); 
plt.savefig("TimeProfile-Comparison.svg", format="svg")
ax.set_title('Time taken for GFF Annotations');
```
