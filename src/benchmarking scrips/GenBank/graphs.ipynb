```python
import seaborn as sns
sns.set_style("whitegrid")
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {'200 Annotations':[152.5,560],'400 Annotations':[152.5,890],'500 Annotations': [152.5,1230]}
df = pd.DataFrame(data)
fig, ax = plt.subplots() 
m = sns.lineplot(df,ax=ax,markers=True,linewidth = 3)
ax.set_xticks(range(0,2));
plt.legend(title='Memory Usage');  
m.set_xticklabels(['Func Start','Annotate & Query'])
ax.set(xlabel='Operation', ylabel='Memory in MiB')
ax.set_title('Memory Usage for 500 Genbank Annotations with the Old Approach');
plt.savefig("Memory-OldApproach.svg", format="svg")
```

```python
old_data = {'Old Approach':[152.5,1345],'New Approach':[152.5,248.5]}

old_df = pd.DataFrame(old_data)

fig, ax = plt.subplots()
m = sns.lineplot(old_df,ax=ax,markers=True,linewidth = 3)

ax.set_xticks(range(0,2));
plt.legend(title='Memory Usage');  
m.set_xticklabels(['Func Start','Annotate & Query'])
ax.set(xlabel='Operation', ylabel='Memory in MiB')
ax.set_title('Memory Usage for 500 Genbank Annotations');
ax.lines[1].set_linestyle("-")
plt.savefig("Memory-Comparison.svg", format="svg")
```

```python
data = {'Old Approach':[35,600,1200,1800,3600]}
df = pd.DataFrame(data)
fig, ax = plt.subplots()
m = sns.lineplot(df,ax=ax,markers=True,linewidth = 3)
#ax.set_xticks(range(0,8));
plt.legend(title='Time Complexity of Old Approach');  
m.set_xticklabels(['0','10','20','50','100','200','500','750','900','1000'])
ax.set(xlabel='Number of annotations', ylabel='Time taken')
ax.set_title('Time Complexity of Genbank Annotations with Old Approach');
plt.savefig("TimeComplexity-OldApproach.svg", format="svg");
```

```python
data = {'Old Approach':[35,600,1200,1800,3600],'New Approach':[2.16,2.45,2.80,3.56,4.56]}
df = pd.DataFrame(data)
fig, ax = plt.subplots()
m = sns.lineplot(df,ax=ax,markers=True,linewidth = 3)
#ax.set_xticks(range(0,8));
plt.legend(title='Time Complexity');  
m.set_xticklabels(['0','10','20','50','100','200','500','750','900','1000'])
ax.set(xlabel='Number of annotations', ylabel='Time')
ax.set_title('Time Complexity for Genbank Annotations');
ax.lines[1].set_linestyle("-")
plt.savefig("TimeComplexity-Comparison.svg", format="svg")
```

```python
data = {'Time taken':[125],'Operations':['Annotate & Query']}
old_df = pd.DataFrame(data)
fig, ax = plt.subplots()
sns.barplot(data=old_df,x='Operations',y='Time taken',ax=ax);
ax.set_title('Time taken for 250 Genbank Annotations');
plt.legend(title='Old Approach'); 
plt.savefig("TimeProfile-OldApproach.svg", format="svg")
```

```python
data = {'Time taken':[2.56,125],'Operations':['Annotate & Query','Annotate & Query'],'Approach':['New Approach','Old Approach']}
old_df = pd.DataFrame(data)
fig, ax = plt.subplots()
sns.barplot(data=old_df,x='Operations',y='Time taken',hue='Approach',ax=ax);
plt.legend(title='Approaches'); 
plt.savefig("TimeProfile-Comparison.svg", format="svg")
ax.set_title('Time taken for 250 Genbank Annotations');
```
